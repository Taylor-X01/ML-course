```python
# read dataset from file
data=pd.read_csv("data/TP3-dataset.csv")
x=data['temperature'].values
x=x.reshape(-1,1)
y=data['pressure'].values
```

```python
# model training
model1 = PolynomialRegression( degree = 2, learning_rate = 0.1, iterations = 1000 )
model1.fit( x, y )
Y1_pred = model1.predict( x )
# MSE for Polynomial Regression
print('MSE for Polynomial Regression: ',mean_squared_error(y,Y1_pred))
```

```python
# Visualization
fig = plt.figure(figsize=(12,7))
plt.scatter( x, y, color = 'green',label='data' )
plt.plot( x, Y1_pred, color = 'black',label='Prediction on the data' )
plt.title( 'polynomial regression:k=2' )
plt.xlabel( 'temperature' )
plt.ylabel( 'pressure' )
plt.legend()
plt.show()
```

```python
model2 = PolynomialRegression( degree = 3, learning_rate = 0.1, iterations = 1000 )
model2.fit( x, y )
Y2_pred = model2.predict( x )
print('MSE for Polynomial Regression: ',mean_squared_error(y,Y2_pred))
```

```python
# Visualization
fig = plt.figure(figsize=(12,7))
plt.scatter( x, y, color = 'green',label='data' )
plt.plot( x, Y2_pred, color = 'black',label='Prediction on the data' )
plt.title( 'polynomial regression:k=3' )
plt.xlabel( 'temperature' )
plt.ylabel( 'pressure' )
plt.legend()
plt.show()
```

```python
model3 = PolynomialRegression( degree = 4, learning_rate = 0.1, iterations = 1000 )
model3.fit( x, y )
# Prediction on training set
Y3_pred = model3.predict( x )
# MSE for Polynomial Regression
print('MSE for Polynomial Regression:',mean_squared_error(y,Y3_pred))
```

```python
# Visualization
fig = plt.figure(figsize=(12,7))
plt.scatter( x, y, color = 'green',label='data' )
plt.plot( x, Y3_pred, color = 'black',label='Prediction on the data' )
plt.title( 'polynomial regression:k=4' )
plt.xlabel( 'temperature' )
plt.ylabel( 'pressure' )
plt.legend()
plt.show()
```

### Comparative plot for polynomial regression

```python
fig = plt.figure(figsize=(12,7))
plt.scatter( x, y, color = 'green',label='data' )
plt.plot( x, Y1_pred, color = 'black',label='K=2'  )
plt.plot( x, Y2_pred, color = 'red',label='K=3'  )
plt.plot( x, Y3_pred, color = 'blue',label='K=4'  )
plt.title( 'comparative plot for polynomial regression' )
plt.xlabel( 'temperature' )
plt.ylabel( 'pressure' )
plt.legend()
plt.show()
```

```python
mse = []
for i in range (8):
    model = PolynomialRegression( degree = 3, learning_rate = (i+1)/10, iterations = 1000 )
    model.fit( x, y )
    # Prediction on training set
    Y_pred = model.predict( x )
    # MSE for Polynomial Regression
    mse.append(mean_squared_error(y,Y_pred))
    print("for learning rate=",(i+1)/10,'MSE=',mse[-1])

    
print(f"\n------------------------------------------------------------\n--> The optimal learning rate is {(np.argmin(mse)+1)/10} \n\t MSE = {np.min(mse)}")
print("------------------------------------------------------------\n")
```
